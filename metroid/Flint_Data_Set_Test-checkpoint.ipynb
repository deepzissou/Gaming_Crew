```python
median_income = [28834, 57144,62843]
x_axis = np.arange(len(median_income))
```

```python

colors= ["firebrick", "plum", "cornflowerblue"]
fig1, ax1 = plt.subplots(figsize=(14,10))
plt.bar(x_axis, median_income, color=colors, alpha=0.5, align="center")


# Tell matplotlib where we would like to place each of our x axis headers
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Flint", "Michigan", "United States"])

# Sets the x limits of the current chart
plt.xlim(-0.75, len(x_axis)-0.25)

# Sets the y limits of the current chart
plt.ylim(0, max(median_income)+10000)

# Give our chart some labels and a tile
plt.title("Comparisons Of Median Income (Flint vs Michigan vs The United States)", fontsize=20)
plt.xlabel("Median Income from 2015 to 2019", fontsize=17)
plt.ylabel("Median Income In US Dollars", fontsize=17)


plt.show()
```

```python
#Household data and pie chart
#Types of households
labels = ["Married", "Single Mother Household", "Single Father HouseHold", "Non-Family HouseHold"]

#Sizes of each households
sizes =[9349, 11737, 2874, 16513] 
#Because colors are fun but I think that we should stick to (primary) colors for now. But not that boring.
colors = ["crimson", "gold", "deepskyblue", "forestgreen"]

#POP!!!!
explode = (0.01, 0, 0, 0)

#The meat of pie
fig1, ax1 = plt.subplots(figsize=(15, 10))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

#Labels which can be changed
plt.title("Household Demographices for Flint, Michigan", fontsize=20)
plt.xlabel("(40,362 Reported Households)", fontsize=10)
plt.ylabel("Types of Households", fontsize=16)
plt.axis("equal")

plt.show
```
