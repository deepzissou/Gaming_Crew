```python
#Dependencies 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)

#Read CSV
full_dataset = pd.read_csv("QuickFacts Aug-26-2021 (1).csv")
full_dataset

```

```python
dataset = full_dataset[["Fact", "Michigan", "United States", "Flint city, Michigan"]]

dataset
```

```python
education_df = dataset[dataset["Fact"].isin(["High school graduate or higher, percent of persons age 25 years+, 2015-2019","Bachelor's degree or higher, percent of persons age 25 years+, 2015-2019"])]

education_df = education_df.set_index('Fact')

education_df 
```

```python
high_school_degree_percentages = [85.30, 90.8, 88.0]
x_axis = np.arange(len(high_school_degree_percentages))
colors= ["lightsteelblue", "royalblue", "darkblue"]
fig1, ax1 = plt.subplots(figsize=(14,10))
plt.bar(x_axis, high_school_degree_percentages, color=colors, alpha=0.5, align="center")
# Tell matplotlib where we would like to place each of our x axis headers
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Flint", "Michigan", "United States"])
# Sets the x limits of the current chart
plt.xlim(-0.75, len(x_axis)-0.25)
# Sets the y limits of the current chart
plt.ylim(0, max(high_school_degree_percentages)+20)
# Give our chart some labels and a tile
plt.title("% of People with at least a High School Diploma", fontsize=20)
plt.xlabel("Locations", fontsize=17)
plt.ylabel("Percentage", fontsize=17)
plt.show()

plt.savefig('High School Diploma')
```
