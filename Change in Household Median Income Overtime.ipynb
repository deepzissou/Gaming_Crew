```python
#Dependencies 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Read CSV
full_dataset = pd.read_csv("Median Household Income.csv")
full_dataset
```

```python
dataset = full_dataset[["Year", "Household Income by Race", "Geography"]]
dataset
```

```python
geography = dataset[dataset["Geography"].isin(["United States", "Michigan", "Flint, MI"])]
geography
```

```python
geography1 = geography.rename(columns={"Household Income by Race" : "Median Household Income in Dollars"})
geography1
```

```python
US = geography1[geography1["Geography"].isin(["United States"])]
US = US.set_index("Year")
US_plot = US[["Median Household Income in Dollars", "Geography"]]
line = US_plot.plot.line()
```

```python
MICHIGAN = geography1[geography1["Geography"].isin(["Michigan"])]
MICHIGAN = MICHIGAN.set_index("Year")
MICHIGAN_plot = MICHIGAN[["Median Household Income in Dollars", "Geography"]]
line1 = MICHIGAN_plot.plot.line()
```

```python
FLINT = geography1[geography1["Geography"].isin(["Flint, MI"])]
FLINT = FLINT.set_index("Year")
FLINT_plot = FLINT[["Median Household Income in Dollars", "Geography"]]
line2 = FLINT_plot.plot.line()
```
