```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = pd.read_csv('Housing Data.csv')
```

```python
#checking database
pd.set_option('display.float_format', '{:.2f}'.format)
file.head()
```

```python
#separating database into two databses, one for housing data one for family dynamics
#Housingdata
housing = file.iloc[0:5,:]
housing = housing.set_index('Fact')
housing.head()
```

```python
#creating smaller dataframes for flint, michigan, us
flinthousing = pd.DataFrame(housing['Flint city, Michigan'])
flinthousingper = flinthousing.iloc[0,0]
flinthousingper = flinthousingper*100

mihousing = pd.DataFrame(housing['Michigan'])
mihousingper = mihousing.iloc[0,0]
mihousingper = mihousingper*100

ushousing = pd.DataFrame(housing['United States'])
ushousingper = ushousing.iloc[0,0]
ushousingper = ushousingper*100

smallhousedf = flinthousingper, mihousingper, ushousingper

smallhousedf
```

```python
loc = ('Flint','Michigan','US')

fig, ax = plt.subplots(figsize=(6,6))
plt.ylabel('Percentage',fontweight='bold',fontsize='16')
plt.xlabel('Location',fontweight='bold',fontsize='16')
plt.title('Owner Occupied Housing Percentage',fontweight='bold',fontsize='16')
ax.set_facecolor('cornflowerblue')
ax.grid(color = 'lightgrey',zorder=0,linewidth=2)
plt.bar(loc,smallhousedf,zorder=3,color='firebrick',edgecolor='black')
plt.show()
```

```python

flint = housing['Flint city, Michigan']
flintmv = flint[flint.index[1]].tolist()

michigan = housing['Michigan']
mimv = michigan[michigan.index[1]].tolist()

us = housing['United States']
usmv = us[us.index[1]].tolist()

mvhousing = flintmv, mimv, usmv

fig, ax = plt.subplots(figsize=(6,6))

loc = ('Flint','Michigan','US')

plt.ylabel('Value of Home',fontweight='bold',fontsize='16')
plt.xlabel('Location', fontweight='bold',fontsize='16')
plt.title('Median Value of Owner Occupied Homes', fontweight='bold',fontsize='16')
ax.set_facecolor('cornflowerblue')
ax.grid(color = 'lightgrey',zorder=0,linewidth=2)
ax.bar(loc,mvhousing, color = 'firebrick',zorder=3,edgecolor='black')
plt.show()
```

```python

mrhousing =  flint[2],michigan[2],us[2],flint[4],michigan[4],us[4]

fig, ax = plt.subplots(figsize=(6,6))

loc = ('Flint Mortgage','Michigan Mortgage','US Mortgage','Flint Rent','Michigan Rent','US Rent')

plt.ylabel('Amount Spent per Month',fontweight='bold',fontsize='16')
plt.xlabel('Mortgage vs Rent by Location',fontweight='bold',fontsize='16')
plt.title('Median Value of Mortgage and Rent costs per Month',fontweight='bold',fontsize='16')
ax.set_facecolor('cornflowerblue')
ax.grid(color = 'lightgrey',zorder=0,linewidth=2)
plt.xticks(rotation = 45)
plt.bar(loc,mrhousing,zorder=3,color='firebrick',edgecolor='black')
plt.show()
```

```python
#separating database into two databses, one for housing data one for family dynamics
#family 
family = file.iloc[8:28,:]
family = family.set_index('Fact')
family
```

```python
familydf = family.groupby(family.index).sum()
familydf
```

```python
#separating by location, male, female
flintfemale = familydf.loc['Female:','Flint city, Michigan']
flintmale = familydf.loc['Male:','Flint city, Michigan']
mifemale = familydf.loc['Female:','Michigan']
mimale = familydf.loc['Male:','Michigan']
usfemale = familydf.loc['Female:','United States']
usmale = familydf.loc['Male:','United States']
```

```python
#total population from separated datasets above 
flintpop = flintfemale + flintmale
uspop = usfemale + usmale
mipop = mifemale+ mimale
print(flintpop, uspop, mipop)
```

```python
#flint male vs female pie plot
flintmale = int(flintmale)
flintfemale = int(flintfemale)

flintpie = (flintmale, flintfemale)
fpielabel = ("Male", "Female")

fig, ax = plt.subplots(figsize=(6,6))
explode = (0,.05)
patches, texts, pcts = ax.pie(flintpie, shadow = True, explode = explode,autopct='%.2f%%',labels=fpielabel)
plt.title('Percentage of Men and Women in Flint',fontweight='bold',fontsize='16')
plt.setp(pcts,color='white', fontweight='bold',fontsize=15)
plt.setp(texts, fontsize=14, fontweight='600')

for i, patch in enumerate(patches):
  texts[i].set_color(patch.get_facecolor())

plt.show()
```

```python
#Michigan male vs female pie plot
mimale = int(mimale)
miemale = int(mifemale)

mipie = (mimale, mifemale)
mipielabel = ("Male", "Female")

fig, ax = plt.subplots(figsize=(6,6))
explode = (0,.05)
patches, texts, pcts = ax.pie(mipie, shadow = True, explode = explode,autopct='%.2f%%',labels=mipielabel)
plt.title('Percentage of Men and Women in Michigan',fontweight='bold',fontsize='16')
plt.setp(pcts,color='white', fontweight='bold',fontsize=15)
plt.setp(texts, fontsize=14, fontweight='600')

for i, patch in enumerate(patches):
  texts[i].set_color(patch.get_facecolor())

plt.show()
```

```python
#US male vs female pie plot
usmale = int(usmale)
usfemale = int(usfemale)

uspie = (usmale, usfemale)
uspielabel = ("Male", "Female")

fig, ax = plt.subplots(figsize=(6,6))
explode = (0,.05)
patches, texts, pcts = ax.pie(uspie, shadow = True, explode = explode,autopct='%.2f%%',labels=uspielabel)
plt.title('Percentage of Men and Women in US',fontweight='bold',fontsize='16')
plt.setp(pcts,color='white', fontweight='bold',fontsize=15)
plt.setp(texts, fontsize=14, fontweight='600')

for i, patch in enumerate(patches):
  texts[i].set_color(patch.get_facecolor())

plt.show()
```
